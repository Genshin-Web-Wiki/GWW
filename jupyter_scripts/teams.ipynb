```python
import json
```

```python
with open('../data/characters_premium.json', 'r') as f:
    premium_team = json.load(f)
```

```python
ids = []
for character in premium_team:
    ids.append(character["id"])
len(ids)
```

```python
for character in premium_team:
    character["premium_team_ids"] = [s.lower().replace(' ', '-') for s in character["premium_team"]]
premium_team
```

```python
ids_team = []
for character in premium_team:
    for char in character['premium_team_ids']:
        ids_team.append(char)
ids_team = set(ids_team)
ids_team
```

```python
# Find items that are only in ids
only_in_ids = set(ids) - set(ids_team)
print("Items only in ids:", only_in_ids)

# Find items that are only in ids_team2
only_in_ids_team = set(ids_team) - set(ids)
print("Items only in ids_team:", only_in_ids_team)
```

```python
for character in premium_team:
    for i, char in enumerate(character['premium_team_ids']):
        if char == 'childe-(tartaglia)':
            character['premium_team_ids'][i] = 'tartaglia'
        elif char == 'childe':
            character['premium_team_ids'][i] = 'tartaglia'
        elif char == 'itto':
            character['premium_team_ids'][i] = 'arataki-itto'
        elif char == 'ayaka':
            character['premium_team_ids'][i] = 'kamisato-ayaka'
        elif char == 'kazuha':
            character['premium_team_ids'][i] = 'kaedehara-kazuha'
        elif char == 'kokomi':
            character['premium_team_ids'][i] = 'sangonomiya-kokomi'
```

```python
ids_team = []
for character in premium_team:
    for char in character['premium_team_ids']:
        ids_team.append(char)
ids_team = set(ids_team)
ids_team
```

```python
# Find items that are only in ids
only_in_ids = set(ids) - set(ids_team)
print("Items only in ids:", only_in_ids)

# Find items that are only in ids_team2
only_in_ids_team = set(ids_team) - set(ids)
print("Items only in ids_team:", only_in_ids_team)
```
